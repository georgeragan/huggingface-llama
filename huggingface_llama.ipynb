```python
import torch
import json
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig,
    pipeline

)
```

HUGGING FACE ACCCOUNT CONFIGURATION

```python
config_data=json.load(open('config.json'))
HF_TOKEN=config_data['HF_TOKEN']
```

```python
model_name="meta-llama/Meta-Llama-3.1-8B-Instruct"
```

QUANTISATION--SHRINK THE MODEL(change the precision of weight)

```python
bnb_config=BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

LOADING THE TOKENIZER AND LLM (Tokenizer--convert raw data into format that model needs)

```python
tokenizer=AutoTokenizer.from_pretrained(
    model_name,
    token=HF_TOKEN
)
tokenizer.pad_token=tokenizer.eos_token
```

```python
model=AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    quantization_config=bnb_config,
    token=HF_TOKEN

)
```

```python
text_generator=pipeline(
    task="text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)
```

```python
def get_response(prompt):
  sequence=text_generator(prompt)
  return sequence[0]['generated_text']
```

```python
prompt="What is the machine learning?"
```

```python
llama_response=get_response(prompt)
```

```python
print(llama_response[len(prompt):])
```
